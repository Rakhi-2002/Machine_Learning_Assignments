```python
#Extracting the Age data 
plt.figure(figsize=(8,6)) #size of figure

labels = 'youngerAge', 'middleAge', 'OldAge' #labels to the pie
sizes = [len(df[df['Age']<50]),len(df[df['Age']==50]),len(df[df['Age']>50])] #sections in the pie chart
colors = ['pink', 'yellowgreen','Skyblue'] #color of the pie
explode = (0,0,0) #Slicing in the pie(whether the parts of the pie having gap in between or not...) 
plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = '%1.2f%%', startangle = 90) #to create the pie
plt.axis('equal')
plt.show #to print(display) the pie
```

```python
#Extracting the Gender data 
#male = len(df[df['Sex']==1])
#female = len(df[df['Sex']==0])

plt.figure(figsize=(8,6))

labels = 'Male', 'Female'
sizes = [len(df[df['Sex']==1]),len(df[df['Sex']==0])]
colors = ['skyblue', 'red']
explode = (0,0) #split the pie into two parts
plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = '%1.2f%%', startangle = 90)
plt.axis('equal')
plt.show
```

```python
#Extracting the data based on Chestpain
plt.figure(figsize=(8,6))
labels = ['ChestPain Type 0', 'ChestPain Type 1','ChestPain Type 2','ChestPain Type 3']

sizes = [len(df[df['ChestPain']=='typical']),
         len(df[df['ChestPain']=='asymptomatic']),
         len(df[df['ChestPain']=='nonanginal']),
         len(df[df['ChestPain']=='nontypical'])]
colors = ['skyblue', 'red', 'green', 'yellow']
explode = (0,0,0,0)
plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = '%1.1f%%', startangle = 180)
plt.axis('equal')
plt.show
```

```python
#Extracting the data based on RestBP
plt.figure(figsize=(8,6))

labels = 'Normal RestBP Level(less than 120/80 mmHg)', 'High RestBP Level(greater than 120/80 mmHg)'
sizes = [len(df[df['RestBP']<120]),len(df[df['RestBP']>120])]
colors = ['blue', 'magenta']
explode = (0,0)
plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = '%1.2f%%', startangle = 90)
plt.axis('equal')
plt.show
```

```python
#Extracting the data based on Cholesterol 
plt.figure(figsize=(8,6))

labels = 'Normal Cholestrol', 'BorderLine High Cholestrol', 'High Cholestrol'
sizes = [len(df[df['Chol']<200]),len(df[df['Chol']<239]),len(df[df['Chol']>240])]
colors = ['pink', 'green','cyan']
explode = (0.1,0.05,0)
plt.title('Pie Chart representing people with various Cholestrol Levels')
plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = '%1.2f%%', startangle = 90)
plt.axis('equal')
plt.show
```

```python
#Splitting the dataset into Training and Testing sets
train, test = train_test_split(df, random_state=0, test_size=0.25)
print("Size of training dataset", train.shape)
print("Size of testing dataset", test.shape)
```

```python
plt.figure(figsize=(8,6))

labels = 'Training Data', 'Testing Data'
sizes = [len(train==1),len(test==0)]
colors = ['lightgreen', 'purple']
explode = (0,0)
plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = '%1.2f%%', startangle = 90)
plt.axis('equal')
plt.show
```
